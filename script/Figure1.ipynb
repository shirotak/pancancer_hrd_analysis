```python
SMALL_SIZE = 7
MEDIUM_SIZE = 8
LARGE_SIZE = 10
BIGGER_SIZE = 12
plt.rc('font', size=LARGE_SIZE)     
plt.rc('axes', titlesize=MEDIUM_SIZE)    
plt.rc('axes', labelsize=MEDIUM_SIZE)    
plt.rc('axes', linewidth=0.5) 
plt.rc('xtick', labelsize=SMALL_SIZE)    
plt.rc('ytick', labelsize=SMALL_SIZE) 
plt.rc('legend', fontsize=MEDIUM_SIZE)    
plt.rc('figure', titlesize=BIGGER_SIZE, dpi=300, figsize=(5,4) )  
```

# Figure 1A

```python
plt_df=pd.read_csv(path_to_data+"Data_for_Figure1.tsv",sep="\t",index_col=0)
value="HRD_score"
plt_df=plt_df.dropna(subset=[value]).sort_values(by=value,ascending=False)
print(len(plt_df))
columns=['gBRCA1_LOH', 'gBRCA2_LOH', 'gHRR_LOH', 'sBRCA1_LOH', 'sBRCA2_LOH','sHRR_LOH']\
+['gBRCA1_nonLOH', 'gBRCA2_nonLOH', 'gHRR_nonLOH', 'sBRCA1_nonLOH','sBRCA2_nonLOH', 'sHRR_nonLOH']\
+['BRCA2_homdel', 'RAD51B_homdel', 'nonBRCA_HRR_homdel']\
+["BRCA1_methylation"]
```

```python
cols=['gBRCA1_LOH', 'gBRCA2_LOH', 'gHRR_LOH', 'sBRCA1_LOH', 'sBRCA2_LOH','sHRR_LOH']
N=plt_df.shape[0]
m=len(cols)

fig = plt.figure(figsize=(15,0.5*m))
for i,col in enumerate(cols):
    ax=fig.add_subplot(m,1,i+1)
    for j in range(N):
        if plt_df[col][j]==1:
            r=patches.Rectangle(xy=(j/N,0),width=1/N,height=1,color="k")
            ax.add_patch(r)
    ax.set_xticks([])
    ax.set_yticks([])
    for axis in ['top','bottom','left','right']:
        ax.spines[axis].set_linewidth(2.5)
    
    # Man-Whitney test
    a,b=plt_df[plt_df[col]==1][value],plt_df[plt_df[col]!=1][value]
    _,p=stats.mannwhitneyu(a,b)
    print(col, " p={:.2e}".format(p) )
        
plt.subplots_adjust(hspace=0)

plt.savefig(path_to_results+"Figure1A_1.png",bbox_inches="tight")
```

```python
cols=['gBRCA1_nonLOH', 'gBRCA2_nonLOH', 'gHRR_nonLOH', 'sBRCA1_nonLOH','sBRCA2_nonLOH', 'sHRR_nonLOH']
N=plt_df.shape[0]
m=len(cols)

fig = plt.figure(figsize=(15,0.5*m))
for i,col in enumerate(cols):
    ax=fig.add_subplot(m,1,i+1)
    for j in range(N):
        if plt_df[col][j]==1:
            r=patches.Rectangle(xy=(j/N,0),width=1/N,height=1,color="k")
            ax.add_patch(r)
    ax.set_xticks([])
    ax.set_yticks([])
    for axis in ['top','bottom','left','right']:
        ax.spines[axis].set_linewidth(2.5)
    
    # Man-Whitney test
    a,b=plt_df[plt_df[col]==1][value],plt_df[plt_df[col]!=1][value]
    _,p=stats.mannwhitneyu(a,b)
    print(col, " p={:.2e}".format(p) )
        
plt.subplots_adjust(hspace=0)

plt.savefig(path_to_results+"Figure1A_2.png",bbox_inches="tight")
```

```python
col="HRD_score"
m=1
fig = plt.figure(figsize=(15,0.5*m))
ax=fig.add_subplot(1,1,1)
ax=sns.heatmap( data=plt_df[[col]].T, cmap="rainbow",xticklabels=0,yticklabels=0, cbar=False,
                   vmin=0,vmax=100)
ax.set_xticks([])
ax.set_yticks([])
for axis in ['top','bottom','left','right']:
    ax.spines[axis].set_linewidth(2.5)
        
plt.subplots_adjust(hspace=0)

plt.savefig(path_to_results+"Figure1A_5.png",bbox_inches="tight")
```

```python
# calculate p_values 
dict_gene_p_diff={ "p-value":{},"mean difference":{} }
for col in columns[0:15]:
    a=plt_df.loc[ plt_df[col]==1,value].values
    b=plt_df.loc[ plt_df[col]==0 ,value].values
    c=np.mean(a )-np.mean(b)
    _,p=stats.mannwhitneyu(a,b)
    dict_gene_p_diff["p-value"][col]="{:.2e}".format(p)
    dict_gene_p_diff["mean difference"][col]=np.round(c,2)

table=pd.DataFrame.from_dict(dict_gene_p_diff)
table["p<0.01"]=( np.array(table["p-value"],dtype=float)<0.01).astype(int)
table["p<1e-4"]=( np.array(table["p-value"],dtype=float)<0.0001).astype(int)
table["p<1e-6"]=( np.array(table["p-value"],dtype=float)<0.000001).astype(int)
table.to_csv(path_to_results+"Figure1A_p_diff_table.tsv",sep="\t")
table
```

```python
# LOH vs nonLOH per alteration
gLOH1=plt_df[ plt_df["gBRCA1_LOH"]==1][value].values
gLOH2=plt_df[ plt_df["gBRCA2_LOH"]==1][value].values
gLOH3=plt_df[ plt_df["gHRR_LOH"]==1][value].values
gnonLOH1=plt_df[ plt_df["gBRCA1_nonLOH"]==1][value].values
gnonLOH2=plt_df[ plt_df["gBRCA2_nonLOH"]==1][value].values
gnonLOH3=plt_df[ plt_df["gHRR_nonLOH"]==1][value].values
sLOH1=plt_df[ plt_df["sBRCA1_LOH"]==1][value].values
sLOH2=plt_df[ plt_df["sBRCA2_LOH"]==1][value].values
sLOH3=plt_df[ plt_df["sHRR_LOH"]==1][value].values
snonLOH1=plt_df[ plt_df["sBRCA1_nonLOH"]==1][value].values
snonLOH2=plt_df[ plt_df["sBRCA2_nonLOH"]==1][value].values
snonLOH3=plt_df[ plt_df["sHRR_nonLOH"]==1][value].values

data_comp=[gLOH1,gnonLOH1,gLOH2,gnonLOH2,gLOH3,gnonLOH3,sLOH1,snonLOH1,sLOH2,snonLOH2,sLOH3,snonLOH3]
```

```python
fig = plt.figure(figsize=(3.5,1.5))
ax = fig.add_subplot(1,1,1)
bp=ax.boxplot(data_comp, showfliers=False,patch_artist=True, 
              widths=0.75)
colors=[(1,0,0,0.7),(0,0,1,0.7)]*6
# box face color
for b, c in zip(bp['boxes'], colors):
    b.set(color=c, linewidth=0.5) 
    b.set_facecolor(c)
    b.set_edgecolor("k")
# median line
for b, c in zip(bp['medians'], colors):
    b.set(color="k",linewidth=0.5)
for b, c in zip(bp['whiskers'], colors*2):
    b.set(linewidth=0.5)
# caps
for b, c in zip(bp['caps'], colors*2):
    b.set( linewidth=0.5)

ax.set_xlabel('')
ax.set_ylabel("HRD score",fontsize=MEDIUM_SIZE)
labels=["g$BRCA1$","g$BRCA2$","gHRR","s$BRCA1$","s$BRCA2$","sHRR"]
ax.set_xticks([1.5,3.5,5.5,7.5,9.5,11.5])
ax.set_xticklabels(labels)
ax.spines['top'].set_visible(False)
ax.spines['right' ].set_visible(False)

plt.savefig(path_to_results+"Figure1A_6.png",bbox_inches="tight")
```
